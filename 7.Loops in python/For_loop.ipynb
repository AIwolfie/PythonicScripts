- In programming, a for loop is a control flow statement that allows you to iterate over a sequence (such as a list, tuple, string, or range) and perform operations on each element in the sequence.

- ## Here's the syntax:

    ```python
   for variable in sequence:
       # Code block to be executed
       # for each element in the sequence

    ```
  - ### Variable:
    Represents each element in the sequence during iteration.

   - ### Sequence:
     The collection of elements over which the loop iterates.

- ##  Example:
  Let's say we have a list of numbers [1, 2, 3, 4, 5], and we want to print each number multiplied by 2 using a for loop:
  ```python
  numbers = [1, 2, 3, 4, 5]

  for num in numbers:
     doubled = num * 2
     print(doubled)
  ```
- ### This will output:
  ```python
  2
  4
  6
  8
  10
  ```

- ### Example Explanation: 
    - Let's break down the example code provided in the 'For Loop' section:
    - ###  Creating a List:
    ```python
    numbers = [1, 2, 3, 4, 5]
    ```
    - Here, we create a list named 'numbers' containing the elements '[1, 2, 3, 4, 5]'.

    - ### For Loop Iteration:
    ```python
    for num in numbers:
       doubled = num * 2
       print(doubled)
    ```
    - The 'for' loop iterates over each element ('num') in the 'numbers' list.
    - For each element, it multiplies the number by 2 ('num * 2') and assigns the result to the variable 'doubled'.
    - It then prints the value of 'doubled', which is the element multiplied by 2.
 
    - ### Output:
      ```python
      2
      4
      6
      8
      10
      ```
      -The output shows each element from the list 'numbers' multiplied by 2, printed on separate lines.

## Usage Tips:
  - ### Iterating Over Lists:
    Use a for loop to iterate over elements in a list and perform operations like calculations, modifications, or printing.
  - ### Looping Through Strings:
    You can also loop through characters in a string using a for loop, treating each character as an element in the sequence.
  - ### Using Range Function: 
    The 'range()' function is often used with for loops to generate a sequence of numbers based on start, stop, and step parameters.
