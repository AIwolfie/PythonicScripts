- Type casting is the process of converting one data type into another.
- Python provides several built-in functions for type casting.

### Example 1: Integer to Float

```python
num_int = 10
num_float = float(num_int)
print("Integer to Float:", num_float)
```
- This code snippet demonstrates converting an integer ('num_int') to a float ('num_float') using the 'float()' function. The output will be '10.0', showing the result of the type casting from integer to float.

### Example 2: Float to Integer

```python
num_float = 3.14
num_int = int(num_float)
print("Float to Integer:", num_int)
```
- Here, a float value ('num_float') is converted to an integer ('num_int') using the 'int()' function. The output will be '3', as the decimal part is truncated when converting from float to integer.

### Example 3: String to Integer

```python
str_num = "25"
num_int = int(str_num)
print("String to Integer:", num_int)
```
- In this example, a string containing a number ('str_num') is converted to an integer using the 'int()' function. The output will be '25', demonstrating how Python can convert strings representing numbers to their corresponding integer values.

### Example 4: Integer to String

```python
num_int = 42
str_num = str(num_int)
print("Integer to String:", str_num)
```
- Here, an integer ('num_int') is converted to a string ('str_num') using the 'str()' function. The output will be "42", showing that the integer value has been converted to its string representation.

### Example 5: String to Float

```python
str_float = "3.14"
num_float = float(str_float)
print("String to Float:", num_float)
```
- This code snippet converts a string representing a float ('str_float') to an actual float value ('num_float') using the 'float()' function. The output will be '3.14', demonstrating the conversion from string to float.

### Example 6: Boolean to Integer

```python
bool_val = True
num_int = int(bool_val)
print("Boolean to Integer:", num_int)
```
- Here, a boolean value ('bool_val') of True is converted to an integer using the 'int()' function. The output will be '1', as True in Python is equivalent to '1' when converted to an integer.

### Example 7: Integer to Boolean

```python
num_int = 0
bool_val = bool(num_int)
print("Integer to Boolean:", bool_val)
```
- This code converts an integer value ('num_int') of '0' to a boolean value ('bool_val') using the 'bool()' function. The output will be False, as '0' in Python is equivalent to False when converted to a boolean.

### Example 8: List to Tuple

```python
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print("List to Tuple:", my_tuple)
```
- Here, a list ('my_list') containing elements '[1, 2, 3]' is converted to a tuple ('my_tuple') using the 'tuple()' function. The output will be '(1, 2, 3)', showing the conversion from list to tuple.

### Example 9: Tuple to List

```python
my_tuple = (4, 5, 6) 
my_list = list(my_tuple)
print("Tuple to List:", my_list)
```
- This code converts a tuple ('my_tuple') containing elements    '(4, 5, 6)' to a list ('my_list') using the 'list()' function. The output will be '[4, 5, 6]', demonstrating the conversion from tuple to list.

### Example 10: Dictionary to List of Keys

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
keys_list = list(my_dict.keys())
print("Dictionary Keys to List:", keys_list)
```
- Finally, this example converts a dictionary's keys ('my_dict.keys()') to a list ('keys_list') using the 'list()' function. The output will be '['a', 'b', 'c']', showing the conversion from dictionary keys to a list of keys.
